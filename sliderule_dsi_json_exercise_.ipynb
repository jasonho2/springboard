## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# import necessary packages

import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
# read the JSON file to a DataFrame

df = pd.read_json('data/world_bank_projects.json')
```

```python
df.info()
```

```python
df[['countryname', 'project_name']].count()
```

```python
df[['countryname', 'project_name']].nunique()
```

```python
# 1. Find the 10 countries with most projects
# group by the number of times the country appears in the DataFrame for any project, 
# printing the countries with 10 most number of projects

countries = df.groupby('countryname')['project_name'].count().sort_values(ascending=False)
print(countries.head(10))
```

```python
# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

themes = json.load((open('data/world_bank_projects.json')))
df1 = json_normalize(themes, 'mjtheme_namecode', ['id'])
df1 # return the entire dataframe
```

```python
# index by 'code' 

df1 = df1.set_index('code')
df1.head()
```

```python
# count the number of times a theme appears for name, printing the 10 most occurring project themes

df1.name.value_counts().head(10)

# notice the blank project theme
```

```python
# 3. In 2. above you will notice that some entries have only the code and the name is missing. 
#    Create a dataframe with the missing names filled in.

# load as dataframe, normalize json to flat file
themes = json.load((open('data/world_bank_projects.json')))
df2 = json_normalize(themes, 'mjtheme_namecode', ['id'])
```

```python
df2.info()
```

```python
# check number of values for each code (notice difference from previous question, top 10 themes) this means there are blanks

df2['code'].value_counts()
```

```python
df2.code.nunique()
```

```python
unique_themes = df2[['name','code']].drop_duplicates()
# drop duplicate name/code combinations

unique_themes = unique_themes.reset_index()
unique_themes = unique_themes.drop('index',axis=1)
# clear all rows where name = ''
unique_themes = unique_themes[unique_themes.name != '']
unique_themes # 11 codes with the name of the theme
```

```python
# merge df2 and unique_themes on code. this will create a dataframe with full code/name rows with no missing values

full_themes = pd.merge(df2, unique_themes, on='code', suffixes=('_x', '')).drop('name_x', axis=1)
full_themes
```

```python
# return count of each theme and notice difference from #2. missing values filled

full_themes['name'].value_counts()
```
